```python
from sklearn.datasets import make_blobs,make_moons,make_circles
from matplotlib import pyplot
from pandas import DataFrame
import seaborn as sns 
import pandas as pd
from sklearn.metrics import silhouette_score
from sklearn.cluster import Birch
sns.set()
```

```python
x,y=make_blobs(n_samples=1000,centers=5,n_features=3)
df = DataFrame(dict(x=x[:,0], y=x[:,1],z=x[:,2], label=y))
colors = {0:'red', 1:'blue', 2:'green',3:'Yellow',4:'orange'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.show()
```

```python
brc = Birch(branching_factor=50, n_clusters=None, threshold=3)
brc.fit(x)
labels=brc.predict(x)
labels
```

```python
pyplot.scatter(x[:,0], x[:,1], c=labels, cmap='rainbow', alpha=0.7, edgecolors='b')
```

```python
brc = Birch(branching_factor=50, n_clusters=None, threshold=5)
brc.fit(x)
labels1=brc.predict(x)
labels1
```

```python
pyplot.scatter(x[:,0], x[:,1], c=labels1, cmap='rainbow', alpha=0.7, edgecolors='b')
```

```python
brc = Birch(branching_factor=50, n_clusters=None, threshold=8)
brc.fit(x)
labels2=brc.predict(x)
labels2
```

```python
brc = Birch(branching_factor=50, n_clusters=None, threshold=10)
brc.fit(x)
labels3=brc.predict(x)
labels3
```

```python
k = [ 3, 5, 8, 10] 
silhouette_scores = [] 
silhouette_scores.append( 
        silhouette_score(x, labels))
silhouette_scores.append( 
        silhouette_score(x, labels1))
silhouette_scores.append( 
        silhouette_score(x, labels2))
silhouette_scores.append( 
        silhouette_score(x, labels3))


pyplot.bar(k, silhouette_scores) 
pyplot.xlabel('Number of clusters', fontsize = 20) 
pyplot.ylabel('S(i)', fontsize = 20) 
pyplot.show() 
```
