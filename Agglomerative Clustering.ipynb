```python
from sklearn.datasets import make_blobs,make_moons,make_circles
from matplotlib import pyplot
from pandas import DataFrame
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import scipy.cluster.hierarchy as shc
import seaborn as sns 
import pandas as pd
sns.set()
```

```python
x,y=make_blobs(n_samples=500,centers=3,n_features=2)
df = DataFrame(dict(x=x[:,0], y=x[:,1], label=y))
colors = {0:'red', 1:'blue', 2:'green'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key,color=colors[key])
pyplot.show()
```

```python
pyplot.figure(figsize =(8, 8)) 
pyplot.title('Visualising the data') 
Dendrogram = shc.dendrogram((shc.linkage(x, method ='ward')))
```

```python
# The Optimal number of clusters are two
k=2
ac2=AgglomerativeClustering(n_clusters=2)
pyplot.figure(figsize =(6, 6))
pyplot.scatter(x[:,0],x[:,1],c=ac2.fit_predict(x),cmap='rainbow')
pyplot.show()
```

```python
k=3
ac3=AgglomerativeClustering(n_clusters=3)
pyplot.figure(figsize =(6, 6))
pyplot.scatter(x[:,0],x[:,1],c=ac3.fit_predict(x),cmap='rainbow')
pyplot.show()
```

```python
# we can determine the Optimal numbers of clusters using a mathematical technique
#Here, we will use the Silhouette Scores
```

```python

k = [2, 3, 4, 5, 6] 
silhouette_scores = [] 
silhouette_scores.append( 
        silhouette_score(x, ac2.fit_predict(x)))
silhouette_scores.append( 
        silhouette_score(x, ac3.fit_predict(x)))
silhouette_scores.append( 
        silhouette_score(x, ac4.fit_predict(x)))
silhouette_scores.append( 
        silhouette_score(x, ac5.fit_predict(x)))
silhouette_scores.append( 
        silhouette_score(x, ac6.fit_predict(x)))


pyplot.bar(k, silhouette_scores) 
pyplot.xlabel('Number of clusters', fontsize = 20) 
pyplot.ylabel('S(i)', fontsize = 20) 
pyplot.show() 
```

```python
#Thus, with the help of the silhouette scores, it is concluded that the optimal number of clusters for the given data 
#and clustering technique is 2 or 3.
```
