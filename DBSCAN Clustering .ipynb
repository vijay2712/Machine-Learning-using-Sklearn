```python
from sklearn.datasets import make_blobs,make_moons,make_circles
from matplotlib import pyplot
from pandas import DataFrame
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import scipy.cluster.hierarchy as shc
import seaborn as sns 
import pandas as pd
sns.set()
```

```python
X, y = make_moons(n_samples=500, noise=.1)
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.show()
```

```python
db_default = DBSCAN(eps = 0.09, min_samples = 2).fit(X) 
labels = db_default.labels_
labels
```

```python
pyplot.figure(figsize =(6, 6))
pyplot.scatter(X[:,0],X[:,1],c=labels,cmap='rainbow')
pyplot.show()
```

```python
db_default = DBSCAN(eps = 0.09, min_samples = 3).fit(X) 
labels1 = db_default.labels_
labels1
pyplot.figure(figsize =(6, 6))
pyplot.scatter(X[:,0],X[:,1],c=labels1,cmap='rainbow')
pyplot.show()
```

```python
db_default = DBSCAN(eps = 0.09, min_samples = 4).fit(X) 
labels2 = db_default.labels_
labels2
pyplot.figure(figsize =(6, 6))
pyplot.scatter(X[:,0],X[:,1],c=labels2,cmap='rainbow')
pyplot.show()
```

```python
db_default = DBSCAN(eps = 0.09, min_samples = 5).fit(X) 
labels3 = db_default.labels_
labels3
pyplot.figure(figsize =(6, 6))
pyplot.scatter(X[:,0],X[:,1],c=labels3,cmap='rainbow')
pyplot.show()
```

```python
db_default = DBSCAN(eps = 0.09, min_samples = 6).fit(X) 
labels4 = db_default.labels_
labels4
pyplot.figure(figsize =(6, 6))
pyplot.scatter(X[:,0],X[:,1],c=labels4,cmap='rainbow')
pyplot.show()
```

```python
k = [ 2, 3, 4, 5,6] 
silhouette_scores = [] 
silhouette_scores.append( 
        silhouette_score(X, labels))
silhouette_scores.append( 
        silhouette_score(X, labels1))
silhouette_scores.append( 
        silhouette_score(X, labels2))
silhouette_scores.append( 
        silhouette_score(X, labels3))
silhouette_scores.append( 
        silhouette_score(X, labels4))

pyplot.bar(k, silhouette_scores) 
pyplot.xlabel('Number of clusters', fontsize = 20) 
pyplot.ylabel('S(i)', fontsize = 20) 
pyplot.show() 
```

```python
#Thus, with the help of the silhouette scores, it is concluded that the optimal number of clusters for the given data 
#and clustering technique is 5
```
